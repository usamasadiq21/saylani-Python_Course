## LINE & BAR PLOT 

### Import Libraries
Seaborn automatically installs
- numpy
- scipy
- pandas
- matplotlib

```python
# Importing Libraires
import seaborn as sns
import matplotlib.pyplot as plt

# Load Dataset from seaborn
phool = sns.load_dataset('iris')
phool

# Draw Line Graph
sns.lineplot(x="sepal_length", y = "sepal_width", data= phool) # Line-plot requires two numeric variables on (x-y) axis.
plt.show()
```

### Adding plot Title

```python
# Adding Graph Title
import seaborn as sns
import matplotlib.pyplot as plt

# Loading Data
flowers_data = sns.load_dataset('iris')
flowers_data

# Making Plot
L_plot = sns.lineplot(x='sepal_length', y='sepal_width', data = flowers_data)
# Title Code
plt.title("Iris Data Plot")
plt.show()
```

## BAR PLOT

```python
# Importing Libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Importing Titanic Data
kashti = sns.load_dataset('titanic')
kashti
```

### Creating Bar-Graph

```python
plt.figure(figsize=(3,4)) # For making chnge in figure size
sns.barplot(x='sex', y='alone', hue='who', data=kashti)
plt.show()
```

In ABove graph we have man on left side and female on right side, if we want to change their positions then we use "Order" attribute.

```python
plt.figure(figsize=(4,4)) # For making chnge in figure size
sns.barplot(x='sex', y='alone', hue='who', order=['female', 'male'], data=kashti)
plt.show()
```
